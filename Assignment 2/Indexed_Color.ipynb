#### Replace image with values from universal color table

```python
from math import sqrt

def __distance(a, b):
    x = (a[0] - b[0]) ** 2
    y = (a[1] - b[1]) ** 2
    z = (a[2] - b[2]) ** 2
    return sqrt(x + y + z)

def compute_nearest(rgb, color_table):
    min_dist = float('inf')
    for color in color_table.values():
        curr_dist = __distance(rgb, color)
        if curr_dist <= min_dist:
            min_dist= curr_dist
            min_color = color

    return tuple(min_color)
    
def transform_to_indexed_colors(image, color_table, title):
    indexed_image = []
    data = image.load()
    for row in range(image.size[0]):
        for col in range(image.size[1]):
            data[row, col] = compute_nearest(data[row, col], color_table)
    #display(image)
    plt.imshow(image)
    plt.title(f"Transformed image with {title} color table")
    plt.show()
    return data
```

## Method 2: Replacing the universal color table by an adaptive colour table

#### Compute adaptive colors with clustering

```python
def generate_adaptive_color_table(image, num_colors):

    # reshape a 3-D datapoint
    pixels = np.reshape(image, (-1, 3))

    # fit the datapoints to 256 clusters
    kmeans = KMeans(n_clusters = num_colors, random_state=0).fit(pixels)
    dominant_colors = kmeans.cluster_centers_.astype(int)

    # choose the cluster centers as the elements of color table
    adaptive_color_table = { k: tuple(color) for k, color in enumerate(dominant_colors)}
    return adaptive_color_table

# load lena image
lena = np.array(Image.open('Lena.png'))
num_colors = 256  

adaptive_color_table = generate_adaptive_color_table(lena, num_colors)

```

#### Display adaptive color table

```python
color_table_img = Image.new("RGB", (16, 16), (0, 0, 0))
pixels = color_table_img.load()
index = 0 
for i in range(16):
    for j in range(16):
        pixels[i,j] = tuple(adaptive_color_table[index])
        index += 1

resized =  color_table_img.resize((256, 256),Image.Resampling.NEAREST)
resized.save("color_table_adaptive.png")

plt.imshow(color_table_img)
plt.title("Adaptive color table 16x16=256")
plt.show()


```
